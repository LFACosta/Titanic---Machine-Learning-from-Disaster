```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
col_drop = ['Name', 'Ticket', 'Fare', 'Cabin']
```

```python
data.drop(columns=col_drop, axis=1, inplace=True)
test.drop(columns=col_drop, axis=1, inplace=True)
```

```python
data.set_index('PassengerId', inplace=True)
test.set_index('PassengerId', inplace=True)
```

```python
data.head()
```

```python
data = pd.get_dummies(data,'dummy', '_',columns=['Sex','Embarked'], drop_first=True)
test = pd.get_dummies(test,'dummy', '_',columns=['Sex','Embarked'], drop_first=True)
```

```python
data.isna().sum()
```

```python
mean_age = data['Age'].mean()
test_mean = test['Age'].mean()
```

```python
data['Age'].fillna(mean_age, inplace=True)
test['Age'].fillna(test_mean, inplace=True)
```

```python
data.head(20)
```

```python
y = data['Survived'].values
X = data.drop('Survived', axis=1).values
```

```python
y = y.reshape(-1,1)
```

```python
regression = LogisticRegression()
```

```python
regression.fit(X,y)
```

```python
predict = regression.predict(test)
```

```python
test.reset_index(inplace=True)
```

```python
id = test['PassengerId'].values
```

```python
result = pd.DataFrame({'PassengerId':id,'Survived':predict}, columns=['PassengerId','Survived'])
```

```python
result.set_index('PassengerId', inplace=True)
```

```python
result.to_csv('Submission_2.csv')
```
